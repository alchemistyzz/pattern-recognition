```python
import numpy as np
mean1=(-1,1)
mean2=(6,2)
mean3=(0.5,7)
cov1=[[1,0.5],[0.5,1]]
cov2=[[1,0],[0,1]]
cov3=[[1,-0.5],[-0.5,1]]
A1 = np.random.multivariate_normal(mean1,cov1,200)
A2 = np.random.multivariate_normal(mean2,cov2,200)
A3 = np.random.multivariate_normal(mean3,cov3,200)
A = np.concatenate((A1,A2,A3),axis=0)
```

```python
A
```

```python
x = A[:,0]
y = A[:,1]
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

model = KMeans()
visualizer = KElbowVisualizer(model,k=(1,7))
visualizer.fit(A)
visualizer.show()
```

```python
kmeans = KMeans(n_clusters=3).fit(A)
colors = ['b', 'g', 'r']
A=A.tolist()
for id,label in enumerate(kmeans.labels_):
    plt.scatter(A[id][0],A[id][1],color =colors[label],marker='o',s=3)
plt.show()
```
