```python
import numpy as np

def pla(X, y):
    """
    感知器学习算法实现
    注意：只能处理线性可分的数据集，输入线性不可分的数据集函数将无法停止
    args:
        X - 训练数据集
        y - 目标标签值
    return:
        w - 权重系数
    """
    done = False
    # 初始化权重系数
    w = np.zeros(X.shape[1])
    print(f'当前的w权重系数为:{w}')
    print(f'='*78)
    # 循环
    while(done == False):
        done = True
        # 遍历训练数据集
        for index in range(len(X)):
            x = X[index]
            print(f'当前使用的样本是{x}')
            print(f'当前的分类结果是:{np.sign(x.dot(w) * y)}')
            # 判定是否与目标值不符
            if x.dot(w) * y[index] <= 0:
                done = False
                # 修正权重系数
                w = w + y[index] * x
            print(f'更正后分类结果是:{np.sign(x.dot(w) * y)}')
            print(f'更新后w权重系数为:{w}') 
            print(f'='*78)
    return w
```

```python
def main():
    x=[
        [4,3,6],
        [2,-2,3],
        [1,0,-3],
        [4,2,3]
       ]
    x=np.array(x)
    y=[-1,1,1,-1]
    y=np.array(y)
    w=pla(x,y)
main()
```
